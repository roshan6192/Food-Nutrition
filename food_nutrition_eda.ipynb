```python
from sklearn.preprocessing import MinMaxScaler

# Combine all dataframes into one (assuming df_group1, df_group2, etc., are already loaded)
df = pd.concat([df_group1, df_group2, df_group3, df_group4, df_group5], ignore_index=True)

# Select key nutrients for nutrition score calculation
nutrient_columns = ['saturated fats', 'monounsaturated fats', 'polyunsaturated fats',
                    'carbohydrates', 'sugars', 'protein', 'dietary fiber', 'cholesterol', 
                    'sodium', 'water', 'vitamin a', 'vitamin b1', 'vitamin b11', 'vitamin b12', 
                    'vitamin b2', 'vitamin b3', 'vitamin b5', 'vitamin b6', 'vitamin c', 
                    'vitamin d', 'vitamin e', 'vitamin k', 'calcium', 'copper', 'iron', 
                    'magnesium', 'manganese', 'phosphorus', 'potassium', 'selenium', 'zinc', 
                    'nutrition density']

# Remove rows with missing values in selected columns
df = df.dropna(subset=nutrient_columns)

# Normalize the selected nutrients (scale between 0 and 1)
scaler = MinMaxScaler()
df[nutrient_columns] = scaler.fit_transform(df[nutrient_columns])

# Define a function to calculate the percentage of calories from each nutrient
def calculate_caloric_contribution(df):
    # Assume typical caloric contributions:
    # Carbohydrates: 4 calories per gram
    # Protein: 4 calories per gram
    # Fat: 9 calories per gram
    
    # Calories from macronutrients
    df['calories_from_carbs'] = df['carbohydrates'] * 4
    df['calories_from_protein'] = df['protein'] * 4
    df['calories_from_fat'] = df['saturated fats'] * 9 + df['monounsaturated fats'] * 9 + df['polyunsaturated fats'] * 9  # Add all fat types
    
    # Total calories (from macronutrients)
    df['total_calories'] = df['calories_from_carbs'] + df['calories_from_protein'] + df['calories_from_fat']
    
    # Avoid division by zero by ensuring total_calories is non-zero
    df['total_calories'] = df['total_calories'].replace(0, 1)  # Replace 0 with 1 to avoid divide by zero
    
    # Calculate percentage of total calories from each macronutrient
    df['percent_calories_from_carbs'] = df['calories_from_carbs'] / df['total_calories']
    df['percent_calories_from_protein'] = df['calories_from_protein'] / df['total_calories']
    df['percent_calories_from_fat'] = df['calories_from_fat'] / df['total_calories']
    
    return df

# Apply the function to calculate the caloric contributions
df = calculate_caloric_contribution(df)

# Define the weights based on % of total calories (you can adjust these based on your preference)
caloric_weights = {
    'carbohydrates': 1.5,  # carbs contribute 1.5 times the weight
    'protein': 2,          # protein is more important
    'fat': 1.2,            # fats are important but slightly less than protein
    'saturated fats': 1.0, # keep as neutral weight
    'monounsaturated fats': 1.0,
    'polyunsaturated fats': 1.0,
    'sugars': 0.5,         # lower weight for sugars
    'dietary fiber': 1.0,  # fiber is important, but less weight than protein
    # (Add other nutrients with their relevant weights here)
}

# Apply caloric weight based on percentage of calories from each nutrient
df['weighted_carbs'] = df['percent_calories_from_carbs'] * caloric_weights['carbohydrates']
df['weighted_protein'] = df['percent_calories_from_protein'] * caloric_weights['protein']
df['weighted_fat'] = df['percent_calories_from_fat'] * caloric_weights['fat']

# Now apply neutral or lower weights to other nutrients
for nutrient in ['saturated fats', 'monounsaturated fats', 'polyunsaturated fats', 'sugars', 
                 'dietary fiber', 'cholesterol', 'sodium', 'water', 'vitamin a', 'vitamin b1', 
                 'vitamin b11', 'vitamin b12', 'vitamin b2', 'vitamin b3', 'vitamin b5', 'vitamin b6', 
                 'vitamin c', 'vitamin d', 'vitamin e', 'vitamin k', 'calcium', 'copper', 'iron', 
                 'magnesium', 'manganese', 'phosphorus', 'potassium', 'selenium', 'zinc', 'nutrition density']:
    df[f'weighted_{nutrient}'] = df[nutrient] * caloric_weights.get(nutrient, 1)

# Calculate the nutrition score (sum of weighted, normalized values)
df['nutrition_score'] = df[[col for col in df.columns if col.startswith('weighted_')]].sum(axis=1)

# Find the most nutritious foods (highest scores)
most_nutritious = df.sort_values(by='nutrition_score', ascending=False).head(10)
print("\nTop 10 Most Nutritious Foods:")
print(most_nutritious[['food', 'nutrition_score']])

# Find the least nutritious foods (lowest scores)
least_nutritious = df.sort_values(by='nutrition_score', ascending=True).head(10)
print("\nTop 10 Least Nutritious Foods:")
print(least_nutritious[['food', 'nutrition_score']])
```

```python
# Assuming df_all contains all combined data (as previously done)
df_all = pd.concat([df_group1, df_group2, df_group3, df_group4, df_group5], ignore_index=True)

# Ensure that the 'food' column exists and rename it to a consistent name
df_all = df_all.rename(columns={'food_column_name_in_each_table': 'food'})  # Adjust this line accordingly

# Nutrients to prioritize (for low sugar and high fiber foods)
prioritize_nutrients = {
    'dietary fiber': 1.0,  # High weight for high dietary fiber
}

# Nutrients to avoid (for low sugar foods)
avoid_nutrients = {
    'sugars': 1.0,  # High weight for avoiding sugars
}

# Clean data: Remove rows with missing values in relevant columns
nutrition_columns = list(prioritize_nutrients.keys()) + list(avoid_nutrients.keys())
df_cleaned = df_all.dropna(subset=nutrition_columns)

# Create a function to calculate nutrition score
def calculate_nutrition_score(row):
    score = 0
    
    # Prioritize nutrients (fiber-rich foods)
    for nutrient, weight in prioritize_nutrients.items():
        score += row[nutrient] * weight
    
    # Subtract for nutrients to avoid (low sugar)
    for nutrient, weight in avoid_nutrients.items():
        score -= row[nutrient] * weight
    
    return score

# Apply nutrition score to each food item
df_cleaned['nutrition_score'] = df_cleaned.apply(calculate_nutrition_score, axis=1)

# Sort the foods by the highest nutrition score
df_suggested = df_cleaned.sort_values(by='nutrition_score', ascending=False)

# Suggest the top foods that are low in sugar and high in fiber
top_suggested_foods = df_suggested[['food', 'nutrition_score', 'dietary fiber', 'sugars']].head(10)

# Display the top 10 foods
print("Top 10 Foods Low in Sugar and High in Dietary Fiber:")
print(top_suggested_foods)
```
